###### if  10000<codekata score<15000   (Reached_expectations.csv)



```python
expected = df.loc[(df['CodeKata Score']>10000) & (df['CodeKata Score']<15000)]
expected.to_csv("Reached_expectations.csv")
```

###### if  7000<codekata score<10000   (Needs_Improvement.csv)


```python
need_improve = df.loc[(df['CodeKata Score']>7000) & (df['CodeKata Score']<10000)]
need_improve.to_csv("Needs_Improvement.csv")
```

###### if  codekate score < 7000        (Unsatisfactory.csv)
